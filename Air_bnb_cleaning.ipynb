```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Yaroslav\Desktop\DataSets\Set25\Airbnb_Open_Data.csv', dtype='string')

df
```

```python
df = df.drop_duplicates()
df = df.drop(columns=[
                        'host_identity_verified', 'lat', 'long',
                        'country', 'country code', 'instant_bookable',
                        'calculated host listings count','house_rules',
                        'license'
                        ])
df = df.rename({'NAME': 'name', 'Construction year': 'construction year'}, axis="columns")

df.head()
```

```python
check_unique = ['neighbourhood group', 'cancellation_policy', 'room type', 'construction year']

for elem in check_unique:
    print(df[elem].unique())
```

```python
def str_replace_func(columns, values_old, values_new):
    for i in range(len(columns)):
        df[columns[i]] = df[columns[i]].str.replace(values_old[i], values_new[i])
    return df

replace_columns = ['price', 'price', 'service fee', 'reviews per month']
values_old = ['$', ',', '$', '.']
values_new = ['', '', '', ',']

df = df.replace({'brookln': 'Brooklyn', 'manhatan': 'Manhattan'})

str_replace_func(replace_columns, values_old, values_new)
```

```python
def fillna_func(columns, values):
    for i in range(len(columns)):
        df[columns[i]] = df[columns[i]].fillna(values[i])
    return df
    
fillna_columns = ['cancellation_policy', 'service fee', 'last review', 'reviews per month', 'review rate number']
fillna_values = ['unknown', '0', 'no review', '0', '0']

fillna_func(fillna_columns, fillna_values)
```

```python
df['neighbourhood group'] = df['neighbourhood group'].fillna(df.groupby('neighbourhood')['neighbourhood group'].transform('last'))

df = df.dropna(subset=['neighbourhood', 'construction year','price', 'minimum nights', 'number of reviews', 'availability 365'])

df
```

```python
df = df.astype({'price': int, 'service fee': int, 'minimum nights': int, 'availability 365': int})
```

```python
df = df[(df['minimum nights'] >= 0) & (df['availability 365'] >= 0)]

df
```

```python
df.to_csv('Air_bnb_PreCleaned_3.csv', index=False)
```
